# Herein, we applied TCGPR to select the best feature subset.

```python
#coding=utf-8
dataSet = "dataset.csv"
initial_set_cap = 2
sampling_cap = 1
# for feature selection
# ratio is recommend as a float <= 0.01
ratio = 0.001
up_search = 100

TCGPR.fit(filePath = dataSet, Mission = 'FEATURE',initial_set_cap = initial_set_cap, sampling_cap = sampling_cap, ratio = ratio, up_search = up_search)
```
