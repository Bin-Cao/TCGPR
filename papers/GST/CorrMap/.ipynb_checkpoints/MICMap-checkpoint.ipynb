```python
###### code of BinCao, MGI, shanghai university, Shanghai
# share for free 
# complie date :2021/9/14 11:22
from minepy import MINE
from pandas._libs import algos as libalgos
import numpy as np
def nolinear_corr(
        self,
        method = "MIC",
        min_periods: int = 1,alpha=0.6, c=15
) :
    """
    Compute pairwise correlation of columns, excluding NA/null values.

    Parameters
    ----------
    method : {'MIC'} or callable
        Method of correlation:

        * MIC : max informal coe
        * callable: callable with input two 1d ndarrays
            and returning a float. Note that the returned matrix from corr
            will have 1 along the diagonals and will be symmetric
            regardless of the callable's behavior.
    min_periods : int, optional
        Minimum number of observations required per pair of columns
        to have a valid result.

    Returns
    -------
    DataFrame
        Correlation matrix.
    Examples
    --------
    >>> df = pd.DataFrame([(.2, .3), (.0, .6), (.6, .0), (.2, .1)],
    ...                   columns=['dogs', 'cats'])
    >>> df.nolinear_corr(method=MIC)
          dogs  cats
    dogs   1.0   0.3
    cats   0.3   1.0
    """
    numeric_df = self._get_numeric_data()
    cols = numeric_df.columns
    idx = cols.copy()
    mat = numeric_df.to_numpy(dtype=float, na_value=np.nan, copy=False)
    
    mine = MINE(alpha=alpha, c=c)
    
    if method == "MIC":
        if min_periods is None:
            min_periods = 1
        mat = mat.T
       
        K = len(cols)
        correl = np.empty((K, K), dtype=float)
        mask = np.isfinite(mat)
        for i, ac in enumerate(mat):
            for j, bc in enumerate(mat):
                if i == j:
                    c = 1.0
                else:
                    ac=pd.DataFrame(ac)
                    bc=pd.DataFrame(bc)
                    mine.compute_score(ac.iloc[:,0], bc.iloc[:,0])
                    c=mine.mic()

                correl[i, j] = c
                correl[j, i] = c
    else:
        raise ValueError(
            "method must be either 'MIC', "
            " or a callable, "
            f"'{method}' was supplied"
        )

    return self._constructor(correl, index=idx, columns=cols)
```
